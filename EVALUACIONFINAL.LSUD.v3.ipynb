```python
import re
class TiendaOnline:
    def __init__(self):
        self.inventario = []  
        self.clientes = {} 
        self.ventas_totales = 0.0 

    def agregar_producto(self, nombre, precio, cantidad):
             
        for producto in self.inventario:
            if producto["nombre".lower()] == nombre.lower():
                producto["cantidad"] += cantidad                
                print(f"El {nombre} ya existía en el inventario. Se ha modificado la cantidad, ha aumentado en {cantidad} unidades")
                break               
            
        else:
            nuevo_producto = {"nombre": nombre, "precio": precio, "cantidad": cantidad}
            self.inventario.append(nuevo_producto)
            print(f"Se ha agregado {nombre} al inventario")
            
            
        
    def ver_inventario(self):
        for producto in self.inventario:
            nombre = producto["nombre"]
            cantidad = producto["cantidad"]
            precio = producto["precio"]
            
            print(f"(Nombre: {nombre}, Precio: {precio}$, Cantidad: {cantidad})")
            

    def buscar_producto(self, nombre):
    
        for producto in self.inventario:
            
            if producto ["nombre".lower()] == nombre.lower():
                cantidad = producto["cantidad"]
                precio = producto["precio"]
                print(f"EL producto {nombre.lower()} está en  el inventario. (Nombre: {nombre.lower()}, Precio: {precio}$, Cantidad: {cantidad})")
                break
        else:
           print(f"El producto {nombre.lower()} no está en el inventario")

    def actualizar_stock(self, nombre, cantidad): 
        for producto in self.inventario:
            if producto ["nombre".lower()]==nombre.lower():
                producto["cantidad"] = cantidad
                print(f"Se ha actualizado la cantidad. El producto {nombre.lower()} tiene actualmente {cantidad} unidades")
                break
                                                  
        else:
            print(f"El producto {nombre.lower()} no está en el inventario y no puede ser modificado")

    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto["nombre".lower()] == nombre.lower():
                self.inventario.remove(producto)
                print(f"Hemos elimnado el producto {nombre.lower()}. El inventario es el siguiente{self.inventario}")
                break
        else:
            print(f"No se puede eliminar el producto {nombre.lower()} porque no existe")

    def calcular_valor_inventario(self):
        total_valor_productos = 0
    
        for producto in self.inventario:
            precio = producto["precio"]
            cantidad = producto["cantidad"]
            valor_producto = precio * cantidad
            total_valor_productos += valor_producto
        print(f"El total del valor del inventario de productos es:{total_valor_productos}$")

    def buscar_producto_regex(self, patron):        
        for producto in self.inventario:        
            if re.findall(patron, producto ["nombre"]):                                                                                                          
                print(f"{producto}")
                
    def realizar_compra(self):
        
        while True:
            carrito = []       
            nombre_producto = input("Ingrese el nombre del producto que desea comprar (o 'terminar' para finalizar la compra): ").lower()

            if nombre_producto.lower() == 'terminar':
                break
            for producto in self.inventario:
                if producto["nombre".lower()] == nombre_producto.lower():
                    
                    cantidad_deseada = int(input(f"Ingrese la cantidad de que desea comprar: "))
                    #cantidad = producto["cantidad"]
                    if cantidad_deseada <= producto["cantidad"]:
                        carrito.append({"nombre": nombre_producto, "precio": producto["precio"], "cantidad": cantidad_deseada})
                        producto["cantidad"] -= cantidad_deseada
                        print(f"El {nombre_producto} se ha agregado al carrito.")
                        
                    else:
                        print(f"No hay suficiente stock de '{nombre_producto.lower()}' en el inventario.")
        
                    total_compra = sum(producto["precio"] * producto["cantidad"] for producto in carrito)
                    print(f"El precio total de la compra es: {total_compra}$")
                    print("Compra realizada con éxito.")
                    return carrito
                           
                                     
                
    def procesar_pago(self):
        try:
            cantidad_total = float(input("¿Cuánto es el total de su compra?"))
            cantidad_pago = float(input("¿Cuanto va a pagar usted?"))
            if cantidad_pago >= cantidad_total:
                cambio= cantidad_pago - cantidad_total
                print(f"el pago se ha realizado correctamente. Su cambio es {cambio} $")
            else:
                print("El importe que va a abonar es menor que el de su compra")
        except ValueError:
            print("Error en el pago. Vuelva a ingresar importes válidos")

    def agregar_cliente(self, nombre, mail):
        cliente = {"nombre": nombre, "mail": mail, "compras": []}
        self.clientes[nombre]= cliente
        print(cliente)

    def ver_clientes(self):
        lista_de_clientes = []
        for nombre, datos_correo in self.clientes.items():
            cliente = {'Nombre': nombre, 'Mail': datos_correo['mail']}
            lista_de_clientes.append(cliente)
        print(lista_de_clientes)    
             
              
              
#-------------------------------------------------
        
    def registrar_compra(self, nombre_cliente, carrito):
            if nombre_cliente in self.clientes:
                total_compra = sum(producto["precio"] * producto["cantidad"] for producto in carrito)
                compra = {"productos": carrito, "total": total_compra}
                
                self.clientes[nombre_cliente]["compras"] = []
                self.clientes[nombre_cliente]["compras"].append(compra)
                self.clientes[nombre_cliente]["ventas_totales"] += total_compra
                print(f"Compra registrada para {nombre_cliente}. Total de la compra: {total_compra}")
            else:
                print(f"No se puede realizar la acción, el cliente '{nombre_cliente}' no está en el diccionario de clientes.")

    def ver_compras_cliente(self, nombre_cliente):
        for cliente in self.clientes:
            if nombre_cliente == cliente["nombre"]:
                print("el cliente está registrado")
            else:
                print("el cliente no está registrado")



    def calcular_ventas_totales(self):
        for cliente in self.clientes:            
            ventas_totales = sum(total_compra)



            
        

   
        















        
                         
        

    
            
             
        
        
```
